```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pickle

with open('eval_metrics_normal_VSM.pkl', 'rb') as f:
    eval_VSM = pickle.load(f)

with open('eval_metrics_google_news.pkl', 'rb') as f:
    eval_Google_News = pickle.load(f)

with open('eval_metrics_bert.pkl', 'rb') as f:
    eval_BERT = pickle.load(f)

with open('eval_metrics_LSA.pkl', 'rb') as f:
    eval_LSA = pickle.load(f)
```

Baseline Model

```python
precisions_VSM = eval_VSM['precisions']
recalls_VSM = eval_VSM['recalls']
fscores_VSM = eval_VSM['fscores']
MAPs_VSM = eval_VSM['MAPs']
nDCGs_VSM = eval_VSM['nDCGs']

plt.plot(range(1, 11), precisions_VSM, c = 'b', label="Precision")
plt.plot(range(1, 11), recalls_VSM, c = 'g', label="Recall")
plt.plot(range(1, 11), fscores_VSM, c = 'r', label="F-Score")
plt.plot(range(1, 11), MAPs_VSM, c = 'c', label="MAP")
plt.plot(range(1, 11), nDCGs_VSM, c = 'm', label="nDCG")
plt.legend()
plt.title("Evaluation Metrics - Vector Space Model")
plt.xlabel("k")
plt.savefig("eval_plot_VSM.png")
```

LSA

```python
precisions_LSA = eval_LSA['precisions']
recalls_LSA = eval_LSA['recalls']
fscores_LSA = eval_LSA['fscores']
MAPs_LSA = eval_LSA['MAPs']
nDCGs_LSA = eval_LSA['nDCGs']

plt.plot(range(1, 11), precisions_LSA, c = 'b', label="Precision")
plt.plot(range(1, 11), recalls_LSA, c = 'g', label="Recall")
plt.plot(range(1, 11), fscores_LSA, c = 'r', label="F-Score")
plt.plot(range(1, 11), MAPs_LSA, c = 'c', label="MAP")
plt.plot(range(1, 11), nDCGs_LSA, c = 'm', label="nDCG")

plt.plot(range(1, 11), precisions_VSM, c = 'b', linestyle = "--")
plt.plot(range(1, 11), recalls_VSM, c = 'g', linestyle = "--")
plt.plot(range(1, 11), fscores_VSM, c = 'r', linestyle = "--")
plt.plot(range(1, 11), MAPs_VSM, c = 'c', linestyle = "--")
plt.plot(range(1, 11), nDCGs_VSM, c = 'm', linestyle = "--")

plt.legend()
plt.title("Evaluation Metrics - LSA vs Benchmark")
plt.xlabel("k")
plt.savefig("eval_plot_LSA.png")
```
